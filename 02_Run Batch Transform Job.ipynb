```python
from sagemaker.transformer import Transformer
import sagemaker

# Define the bucket and S3 input/output paths
bucket = 'sagemaker-us-east-1-877305355343'
input_location = f's3://{bucket}/batch_input/'
output_location = f's3://{bucket}/inference/output/'

# Initialize the transformer with deployed model
transformer = Transformer(
    model_name='FacialExpressionAnalysisAAI540',
    instance_count=1,
    instance_type='ml.t2.medium', 
    output_path=output_location,
    assemble_with='Line',
    accept='application/json'
)


# Start the batch transform job
transformer.transform(
    data=input_location,
    content_type='application/x-image',
    split_type='None'
)

# Wait for job to complete
transformer.wait()
print("✓ Batch Transform completed.")
```
